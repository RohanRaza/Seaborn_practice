```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("housing.csv")
data.head()
```

```python
data.columns
```

```python
sns.relplot(x = "Gr Liv Area", y ='SalePrice', data = data, hue = 'Overall Qual', \
           palette = 'RdYlBu',size = 'Garage Area', sizes = (1,200),style = 'Rooms',\
           col = "Year")



plt.show()
```

```python
sp = data[data["Gr Liv Area"] > 4500]
sp.head()
```

```python
sns.relplot(x = "Gr Liv Area", y ='SalePrice', data = sp, hue = "Foundation")

plt.show()
```

```python
sp.corr()['SalePrice'].sort_values(ascending = False)
```

```python
sns.relplot(x = "Gr Liv Area", y ='SalePrice', data = sp, \
           palette = 'RdYlBu',size = 'Garage Area', sizes = (1,200),style = '2nd Flr SF',\
           col = "Year")



plt.show()
```
